```python
import pandas as pd
from fastai.collab import *
```

```python
reviews = pd.read_csv("./storage/mc_ratings.csv")
```

```python
reviews.head()
```

```python
binary_ratings = [1 if rating >= 50 else 0 for rating in reviews['rating'].values]
reviews['binary_ratings'] = pd.DataFrame(binary_ratings)
```

```python
reviews
```

```python
data = CollabDataBunch.from_df(reviews, user_name="critic_url", item_name="movie_url", rating_name="binary_ratings")
```

```python
learn = collab_learner(data, n_factors=20, y_range=(0.,1.), metrics=root_mean_squared_error)
```

```python
learn.fit_one_cycle(5, 5e-3, wd=0.1)
```

```python
rt_reviews = pd.read_csv("./storage/reviews.csv")
```

```python
rt_reviews
```

```python
rt_binary_ratings = [1 if rating == "fresh" else 0 for rating in rt_reviews['rt_score'].values]
rt_reviews['binary_ratings'] = pd.DataFrame(rt_binary_ratings)
```

```python
rt_reviews
```

```python
rt_counts = dict()
for reviewer_name in rt_reviews['reviewer_id'].values:
    if reviewer_name in rt_counts:
        rt_counts[reviewer_name] += 1
    else:
        rt_counts[reviewer_name] = 1
```

```python
culled_reviews = []
for row in rt_reviews.iterrows():
    row = row[1]
    if rt_counts[row['reviewer_id']] >= 20:
        culled_reviews.append(row)
culled_reviews = pd.DataFrame(culled_reviews)
```

```python
culled_reviews
```

```python
rt_data = CollabDataBunch.from_df(rt_reviews, user_name="reviewer_id", item_name="movie_id", rating_name="binary_ratings")
rt_learn = collab_learner(rt_data, n_factors=20, y_range=(0.,1.), metrics=root_mean_squared_error)
rt_learn.fit_one_cycle(5, 5e-3, wd=0.1)
```
